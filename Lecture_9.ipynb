# List traverser function

```python
def get_report_df(report_list):
    
    # save an empty mega report card data frame 
    mega_report_df=pd.DataFrame()
    #For loop
    #for each class Dictionary in the list :
   ########################################################################################
    for division in report_list:
    
    # make scores df 
        scores_df=pd.DataFrame(division["scores"])
    # make emails df
        
        emails_df=pd.DataFrame(division["emails"])
        emails_df["student_email"] = emails_df["student_email"].str.lower()
    
    # merge the scores df with emails df
        scores_emails_df=pd.merge(scores_df,emails_df, how = "left" , on =["student_name"])
    
    # add the class column
        scores_emails_df["class"]=division["class"]
    
    # add the division column
        scores_emails_df["division"]=division["division"]
    
    # add the teacher column 
        scores_emails_df["teacher"]=division["teacher"]
    
    # add (concat)  it to a mega report card dataframe
        mega_report_df=pd.concat([mega_report_df,scores_emails_df],ignore_index = True)
   ###########################################################################################
   # add total score
    mega_report_df["total_score"] = mega_report_df["math_scores"] +mega_report_df["chemistry_scores"] +mega_report_df["information_technology_scores"]+mega_report_df["physics_scores"]
   # add average scores
    mega_report_df["average_scores"] = mega_report_df["total_score"]/4
    
    # add pass fail column
    mega_report_df.loc[mega_report_df["average_scores"]>=35,"status"]= "Pass"
    mega_report_df.loc[mega_report_df["average_scores"]<35,"status"]= "Fail"
    
    # add character Grades
    mega_report_df.loc[(mega_report_df["average_scores"]>=90),"Grade"] = "A"
    mega_report_df.loc[(mega_report_df["average_scores"]>=70) & (mega_report_df["average_scores"]<90),"Grade"] = "B"
    mega_report_df.loc[(mega_report_df["average_scores"]>=50) & (mega_report_df["average_scores"]<70),"Grade"] = "C"
    mega_report_df.loc[(mega_report_df["average_scores"]>=35) & (mega_report_df["average_scores"]<50),"Grade"] = "D"
    mega_report_df.loc[(mega_report_df["average_scores"]<35),"Grade"] = "F"
    
    #return the mega report card data frame 
    return mega_report_df
```
