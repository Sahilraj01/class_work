- What is the average chemistry grade? 64.2
- What is the average biology grade? 56.0

```python
report_card_df["total_grade"] = report_card_df["physics_grade"]+report_card_df["chemistry_grades"]+report_card_df["math_grades"]+report_card_df["biology_grade"] 
report_card_df
```

```python
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
report_card_df["average_grade"] = report_card_df["total_grade"]/4
report_card_df
```

```python
report_card_insights_df.to_csv("Report Card insights.csv")
```

```python
report_card_df
```

```python
report_card_df.loc[report_card_df["total_grade"]>=140,"status"] = "Pass"
report_card_df.loc[report_card_df["total_grade"]<140,"status"] = "Fail"
report_card_df
```

```python
# Making a grade column
report_card_df.loc[(report_card_df["average_grade"] >=90),"Grade"] = "A"
report_card_df.loc[(report_card_df["average_grade"] >=70) & (report_card_df["average_grade"]<90),"Grade"] = "B"
report_card_df.loc[(report_card_df["average_grade"] >=50) & (report_card_df["average_grade"]<70),"Grade"] = "C"
report_card_df.loc[(report_card_df["average_grade"] >=35) & (report_card_df["average_grade"]<50),"Grade"] = "D"
report_card_df.loc[(report_card_df["average_grade"] <35),"Grade"] = "A"
```
