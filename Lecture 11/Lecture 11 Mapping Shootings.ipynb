```python
# setup
import pandas as pd 
```

```python
#
shootings_df=pd.read_csv("Mass-Shooting-Data-2016.csv")
shootings_df
```

```python
# First ten rows in my data frame
shootings_df.head(10)
```

```python
# average injuries and killed by state and city 
avg_ing_kil_df=shootings_df.groupby(["state","city"]).agg(avg_injuries = ("injured","mean"),
                                                          avg_killed = ("killed","mean"))
avg_ing_kil_df
```

```python
#Total injuries and killed by state and city 
total_inj_kil_df=shootings_df.groupby(["state","city"]).agg(total_injuries = ("injured","sum"),
                                                        total_killed = ("killed","sum"))
total_inj_kil_df
```

```python
# sort by killed column in ascending 
total_inj_kil_df.sort_values("total_killed", ascending= True)
```

```python
# sort by killed column in descending 
total_inj_kil_df.sort_values("total_killed", ascending= False)
```

```python
# top 5 countries with most kills
top_5_killed = total_inj_kil_df.sort_values("total_killed", ascending= False).head(5)
top_5_killed 
```

```python
# Top 5 states with most injuries
total_inj_df=shootings_df.groupby(["state"]).agg(total_injuries = ("injured","sum"))
total_inj_df=total_inj_df.sort_values("total_injuries", ascending= False).head(5)
total_inj_df
```

```python
# Top 5 states with least injuries
total_inj_df=shootings_df.groupby(["state"]).agg(total_injuries = ("injured","sum"))
total_inj_df=total_inj_df.sort_values("total_injuries", ascending= True).head(5)
total_inj_df
```
